```python
#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns
```

```python
#Read csv file
df = pd.read_csv("deaths-in-custody.csv")
```

```python
counts = df.groupby(['race', 'manner_of_death']).size()
print (counts)
```

```python
counts.to_csv("test.csv", encoding='utf-8', index=False)
```
